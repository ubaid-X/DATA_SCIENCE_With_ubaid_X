>> # __Cheapter 1: Numbers and Operations in algebra__

> ### __1. Numbers__

1. __Natural Numbers__: These are the numbers we naturally count with. They start from 1 and go on indefinitely (1, 2, 3, …). 
  - __Example__: The number of apples in a basket (assuming there’s at least one) is a natural number.

2. __Whole Numbers__: Whole numbers include all natural numbers and the number 0. 
  - __Example__: The number of cars in a parking lot, which could be zero or any natural number.

3. __Integers__: Integers are whole numbers that can be positive, negative, or zero. This includes numbers like -3, -2, -1, 0, 1, 2, 3, … 
  - __Example__: The temperature above or below zero on a thermometer is an integer.

4. __Rational Numbers__: Rational numbers are numbers that can be expressed as the quotient of two integers, where the denominator is not zero. They can be written as fractions. 
  - __Example__: The number 1/2 is rational because it represents one divided by two.

5. __Irrational Numbers__: Irrational numbers cannot be expressed as a simple fraction; their decimal expansions are non-repeating and non-terminating. 
  - __Example__: The square root of 2 (√2) is an irrational number because it cannot be precisely written as a fraction.

6. __Real Numbers__: Real numbers include all the rational and irrational numbers. The number line represents real numbers.
  - __Example__: Pi (π), which is approximately 3.14159, is a real number.

---

> ### __2.Operations__

__Arithmetic Operations__: These are the basic operations we perform with numbers:

1. __Addition (+)__: Combining two amounts. 
   - __Example__: If you have 3 apples and get 2 more, you now have 3 + 2 = 5 apples.

2. __Subtraction (-)__: Finding the difference between amounts. 
   - __Example__: If you have 5 apples and eat 2, you have 5 – 2 = 3 apples left.
  
3. __Multiplication (×)__: Repeated addition of the same number. 
   - __Example__: If you have 4 bags with 3 apples each, you have 4 × 3 = 12 apples.
  
4. __Division (÷)__: Splitting into equal parts or groups. 
   - __Example__: If 12 apples are divided among 4 friends, each friend gets 12 ÷ 4 = 3 apples.

---


> ### __3. Order of Operations (PEMDAS/BODMAS)__

This rule helps us remember the sequence to solve operations in a complex expression:

- Parentheses/Brackets
- Exponents/Orders
- Multiplication and Division (from left to right)
- Addition and Subtraction (from left to right) `=>` Example: For 3 + 4 × 2, multiply first (4 × 2 = 8), then add (3 + 8 = 11).

---

> ### __4. Key Terms in Algebra__

1. __Variable__: A variable is a symbol (usually a letter) that stands for a number that can vary or change. 
   - __Example__: In the expression x + 2, x is a variable.

2. __Constant__: A constant is a fixed value that does not change. 
   - __Example__: In the expression x + 2, the number 2 is a constant.

3. __Coefficient__: A coefficient is a number used to multiply a variable.
   -  __Example__: In 3x, 3 is the coefficient of the variable x.

4. __Expression__: An expression is a combination of variables, numbers, and operations. 
    - __Example__: 2x + 3y – 5 is an expression.

5. __Equation__: An equation is a statement that two expressions are equal, indicated by the equal sign (=). 
   - __Example__: 2x + 3 = 11 is an equation.

6. __Inequality__: An inequality is like an equation, but instead of equality, it shows a relationship of less than (<), greater than (>), less than or equal to (≤), or greater than or equal to (≥). 
   - __Example__: x + 3 > 5 is an inequality.
