# Linear Algebra Essentials: Matrices and Systems of Equations

## Overview
This guide covers fundamental linear algebra concepts including matrix inverses, systems of equations, and their matrix representations. We'll explore definitions, solutions, visualizations, and real-world applications.

---

## 1. Inverse of a Matrix

### Definition
The inverse of a square matrix **A**, denoted as **A⁻¹**, is a matrix such that when multiplied by **A** yields the identity matrix:
**A × A⁻¹ = I**

### Step-by-Step Calculation
**For 2×2 matrix:**
If A = $\begin{bmatrix} a & b \\ c & d \end{bmatrix}$, then:
A⁻¹ = $\frac{1}{ad-bc}$ $\begin{bmatrix} d & -b \\ -c & a \end{bmatrix}$

**General steps:**
1. Verify matrix is square
2. Calculate determinant `|A|`
3. Make sure `det(A) ≠ 0` (if it's 0, inverse doesn't exist)
4. Swap the the main diagonal elements `a and d,` and change the sign of  `b and c.`
5. Multiply by `1 / det(A)`

### Example
Find inverse of A = $\begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix}$:

1. det(A) = (2×3) - (1×1) = 5
2. A⁻¹ = $\frac{1}{5}$ $\begin{bmatrix} 3 & -1 \\ -1 & 2 \end{bmatrix}$ = $\begin{bmatrix} 0.6 & -0.2 \\ -0.2 & 0.4 \end{bmatrix}$

### Visualization
```
Original Matrix A    Inverse Matrix A⁻¹    Identity Matrix
[2   1]           [0.6  -0.2]          [1  0]
[1   3]           [-0.2  0.4]           [0  1]
```

### Tips
- Only square matrices can have inverses
- Matrix must be non-singular (det ≠ 0)
- Use calculator/software for matrices larger than 3×3

### Applications
- Cryptography (encoding/decoding)
- Computer graphics transformations
- Solving systems of equations

## 2. System of Equations

### Definition
A set of equations with multiple variables that must all be satisfied simultaneously.

### Types
1. **Consistent Independent**: Exactly one solution (lines intersect at one point)
2. **Consistent Dependent**: Infinite solutions (lines are identical)
3. **Inconsistent**: No solution (lines are parallel)

### Example
**System:**
- 2x + y = 5
- x - y = 1

**Solution:** (x = 2, y = 1)

### Visualization
```
y-axis
  |
5 |  2x+y=5
  |   /
  |  /
  | /
  |/ 
2 +---x-axis
  | 1 2
  |
  | x-y=1
```

### Tips
- Number equations to track steps
- Check solutions in all original equations
- Graph to visualize solutions

## 3. Types of Linear Equations

### Two Variables
General form: ax + by = c

**Example:**
3x + 2y = 6

**Graph:** Straight line in 2D plane

### Three Variables
General form: ax + by + cz = d

**Example:**
2x + y - 3z = 5

**Graph:** Plane in 3D space

### Higher Dimensions
General form: a₁x₁ + a₂x₂ + ... + aₙxₙ = b

**Applications:**
- Economics (input-output models)
- Engineering (circuit analysis)
- Data science (linear regression)

## 4. Writing System of Equations

### Step-by-Step Process
1. Identify variables and assign letters
2. Write equations based on problem statements
3. Ensure consistent variable order
4. Include all constraints

### Example
**Problem:** Movie tickets cost $8 adults, $5 children. Total 12 tickets sold for $81.

**Variables:**
- Let a = number of adult tickets
- Let c = number of child tickets

**Equations:**
1. a + c = 12  (total tickets)
2. 8a + 5c = 81 (total cost)

### Tips
- Clearly define variables
- Write units when applicable
- Double-check relationships

## 5. Matrix Form of System of Equations

### Conversion Process
A system of equations can be written as **AX = B** where:
- **A** = Coefficient matrix
- **X** = Variable matrix
- **B** = Constant matrix

### Example
**System:**
- 2x + 3y = 8
- x - y = 1

**Matrix form:**
$\begin{bmatrix} 2 & 3 \\ 1 & -1 \end{bmatrix}$ $\begin{bmatrix} x \\ y \end{bmatrix}$ = $\begin{bmatrix} 8 \\ 1 \end{bmatrix}$

### Solution Using Inverse
If AX = B, then X = A⁻¹B

### Visualization
```
Coefficient Matrix A   Variable Matrix X   Constant Matrix B
[2   3 ]             [x]                [8]
[1  -1 ]             [y]                [1]
```

### Advantages
- Compact representation
- Efficient computational solution
- Scalable to large systems

### Limitations
- Only works for linear systems
- Requires coefficient matrix to be invertible

# Solving Systems of Linear Equations: Methods and Applications

## Overview
This guide covers the six main methods for solving systems of linear equations, including their step-by-step processes, examples, advantages, limitations, and ideal use cases.

---

## 1. Graphical Method

### Definition
Solving a system of equations by plotting each equation on a graph and identifying intersection points.

### Step-by-Step Process
1. Rewrite each equation in slope-intercept form (y = mx + b)
2. Plot each line on the same coordinate plane
3. Identify the intersection point(s)
4. Verify the solution by substituting back into original equations

### Example
**System:**
- y = 2x + 1
- y = -x + 4

**Solution:**
1. Plot both lines
2. Find intersection at (1, 3)
3. Verify: 3 = 2(1)+1 ✓ and 3 = -(1)+4 ✓

### Visualization
```
y-axis
  |
5 |   y = -x + 4
  |    /
  |   /
  |  / 
  | / 
3 +---• (1,3)
  |/   
2 |    y = 2x + 1
  |    
  +---+---+--- x-axis
    1 2 3 4
```

### Advantages
- Visual and intuitive
- Good for understanding solutions geometrically

### Limitations
- Inaccurate for non-integer solutions
- Impractical for 3+ variables
- Difficult to parallel or coincident lines

### Use Cases
- Quick visualization of 2-variable systems
- Educational purposes to understand solutions

---

## 2. Substitution Method

### Definition
Solving by isolating one variable and substituting its expression into the other equation.

### Step-by-Step Process
1. Solve one equation for one variable
2. Substitute this expression into the other equation
3. Solve for the remaining variable
4. Substitute back to find the first variable
5. Verify the solution

### Example
**System:**
- 2x + y = 5
- x - y = 1

**Solution:**
1. Solve equation 2 for x: x = y + 1
2. Substitute into equation 1: 2(y+1) + y = 5
3. Simplify: 2y + 2 + y = 5 → 3y = 3 → y = 1
4. Substitute back: x = 1 + 1 = 2
5. Verify: 2(2)+1=5 ✓ and 2-1=1 ✓

### Advantages
- Works for any system with exact solutions
- Straightforward for 2-variable systems
- No special tools required

### Limitations
- Can get messy with fractions
- Less efficient for 3+ variables
- Not ideal for computer implementation

### Use Cases
- Small systems (2-3 variables)
- When one variable is easily isolated
- Algebraic solutions without matrices

---

## 3. Elimination Method

### Definition
Solving by adding or subtracting equations to eliminate variables systematically.

### Step-by-Step Process
1. Align equations with like terms in columns
2. Multiply equations to make coefficients match
3. Add or subtract equations to eliminate one variable
4. Solve for the remaining variable
5. Substitute back to find other variables
6. Verify the solution

### Example
**System:**
- 3x + 2y = 4
- 2x - y = 3

**Solution:**
1. Multiply equation 2 by 2: 4x - 2y = 6
2. Add to equation 1: (3x+2y) + (4x-2y) = 4+6 → 7x = 10 → x = 10/7
3. Substitute: 2(10/7) - y = 3 → 20/7 - y = 3 → y = 20/7 - 21/7 = -1/7
4. Verify: 3(10/7)+2(-1/7)=30/7-2/7=28/7=4 ✓

### Advantages
- Works well for integer coefficients
- Systematic approach
- Good for 2-3 variable systems

### Limitations
- Can involve fractions
- Becomes cumbersome with many variables
- Requires careful arithmetic

### Use Cases
- Systems with integer coefficients
- When variables can be easily eliminated
- Manual calculations for small systems

---

## 4. Matrix Method (Inverse Matrix)

### Definition
Solving using matrix algebra by expressing the system as AX = B and calculating X = A⁻¹B.

### Step-by-Step Process
1. Write system in matrix form: AX = B
2. Verify A is invertible (det ≠ 0)
3. Calculate A⁻¹
4. Multiply: X = A⁻¹B
5. Verify solution

### Example
**System:**
- 2x + y = 5
- x - y = 1

**Solution:**
1. Matrix form: $\begin{bmatrix}2 & 1\\1 & -1\end{bmatrix}$ $\begin{bmatrix}x\\y\end{bmatrix}$ = $\begin{bmatrix}5\\1\end{bmatrix}$
2. det(A) = (2)(-1) - (1)(1) = -3 ≠ 0
3. A⁻¹ = $\frac{1}{-3}$ $\begin{bmatrix}-1 & -1\\-1 & 2\end{bmatrix}$ = $\begin{bmatrix}1/3 & 1/3\\1/3 & -2/3\end{bmatrix}$
4. X = A⁻¹B = $\begin{bmatrix}1/3 & 1/3\\1/3 & -2/3\end{bmatrix}$ $\begin{bmatrix}5\\1\end{bmatrix}$ = $\begin{bmatrix}2\\1\end{bmatrix}$
5. Verify: 2(2)+1=5 ✓ and 2-1=1 ✓

### Advantages
- Elegant mathematical approach
- Direct solution method
- Works for any number of variables (theoretically)

### Limitations
- Requires matrix inversion
- Only works for invertible matrices
- Computationally expensive for large systems

### Use Cases
- Theoretical mathematics
- Small to medium systems with invertible matrices
- When working with matrix operations already

---

## 5. Gaussian Elimination

### Definition
Solving by converting the system to row-echelon form through elementary row operations.

### Step-by-Step Process
1. Write the augmented matrix [A|B]
2. Use row operations to create upper triangular form
3. Perform back substitution to find solutions
4. Verify solution

### Example
**System:**
- 2x + y - z = 3
- x - y + z = 2
- 3x + 2y + z = 4

**Solution:**
1. Augmented matrix:
   $\left[\begin{array}{ccc|c}
   2 & 1 & -1 & 3\\
   1 & -1 & 1 & 2\\
   3 & 2 & 1 & 4
   \end{array}\right]$

2. Row operations to upper triangular form:
   R2 ← 2R2 - R1: $\left[\begin{array}{ccc|c}
   2 & 1 & -1 & 3\\
   0 & -3 & 3 & 1\\
   3 & 2 & 1 & 4
   \end{array}\right]$
   
   R3 ← 2R3 - 3R1: $\left[\begin{array}{ccc|c}
   2 & 1 & -1 & 3\\
   0 & -3 & 3 & 1\\
   0 & 1 & 5 & -1
   \end{array}\right]$
   
   (Continue until upper triangular...)

3. Back substitution to find z, then y, then x

### Advantages
- Works for any system (consistent or inconsistent)
- Efficient for computer implementation
- Can handle non-invertible matrices

### Limitations
- Requires careful bookkeeping
- Potential for numerical instability
- Many steps for manual calculation

### Use Cases
- Computer algorithms (most common method)
- Large systems of equations
- Systems where matrix may not be invertible

## 6. Cramer's Rule

### Definition
Solving using determinants of matrices derived from the coefficient matrix.

### Step-by-Step Process
1. Verify det(A) ≠ 0
2. For each variable xᵢ, create matrix Aᵢ by replacing column i with B
3. Calculate xᵢ = det(Aᵢ) / det(A)
4. Verify solution

### Example
**System:**
- 2x + y = 5
- x - y = 1

**Solution:**
1. det(A) = (2)(-1) - (1)(1) = -3
2. For x: Aₓ = $\begin{bmatrix}5 & 1\\1 & -1\end{bmatrix}$, det(Aₓ) = (5)(-1) - (1)(1) = -6
3. For y: Aᵧ = $\begin{bmatrix}2 & 5\\1 & 1\end{bmatrix}$, det(Aᵧ) = (2)(1) - (5)(1) = -3
4. x = det(Aₓ)/det(A) = (-6)/(-3) = 2
5. y = det(Aᵧ)/det(A) = (-3)/(-3) = 1
6. Verify: 2(2)+1=5 ✓ and 2-1=1 ✓

### Advantages
- Direct formula for each variable
- Elegant for theoretical purposes
- Works for any invertible matrix

### Limitations
- Computationally expensive for large systems
- Requires calculating many determinants
- Only works when det(A) ≠ 0

### Use Cases
- Small systems (2-3 variables)
- Theoretical applications
- When needing specific variables without solving entire system

## Method Selection Guide

| Method | Best For | Avoid When |
|--------|----------|------------|
| Graphical | Visualization, 2 variables | Precision needed, 3+ variables |
| Substitution | Small systems, easy isolation | Many variables, complicated coefficients |
| Elimination | Integer coefficients, small systems | Many variables, fractions likely |
| Matrix Inverse | Theoretical work, small matrices | Large systems, non-invertible matrices |
| Gaussian Elimination | Computer solutions, large systems | Manual calculations, very small systems |
| Cramer's Rule | Small systems, theoretical use | Large systems, when det(A) ≈ 0 |
