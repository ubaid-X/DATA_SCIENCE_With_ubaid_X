### __1. Normal Distribution__

````python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducible results
np.random.seed(42)

# Generate sample from normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation
sample_size = 10000
data = np.random.normal(loc=mu, scale=sigma, size=sample_size)

# Create histogram with density plot
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data, 
                           bins=30, 
                           density=True, 
                           alpha=0.7, 
                           color='skyblue',
                           edgecolor='black',
                           label='Sample Distribution')

# Overlay theoretical normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'r-', linewidth=2, label='Theoretical N(0,1)')

# Add labels and formatting
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normally Distributed Data: Sample vs. Theoretical Distribution')
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Verify distribution quality
print(f"Sample mean: {np.mean(data):.4f} (Target: {mu})")
print(f"Sample std: {np.std(data, ddof=1):.4f} (Target: {sigma})")
print(f"Sample size: {sample_size}")
````

### __2. Uniform Distribution__

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate data from a perfect uniform distribution
np.random.seed(42)  # Ensures reproducible results
low = 0
high = 10
sample_size = 1000
data = np.random.uniform(low=low, high=high, size=sample_size) # :cite[1]:cite[8]

# 2. Plot the histogram
plt.hist(data, bins=15, density=True, alpha=0.7, edgecolor='black', color='skyblue') # :cite[5]
plt.title('Histogram of a Perfect Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
# Overlay the theoretical uniform density function
plt.axhline(y=1/(high-low), color='red', linestyle='--', linewidth=2, label=f'Theoretical PDF (1/({high}-{low}))') # :cite[1]
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 3. Simple Statistical Verification
# Calculate theoretical and sample statistics
theoretical_mean = (low + high) / 2
theoretical_std = np.sqrt((high - low)**2 / 12) # :cite[6]:cite[9]

sample_mean = np.mean(data)
sample_std = np.std(data)

print("Statistical Verification Results:")
print(f"Theoretical Mean: {theoretical_mean:.4f}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Theoretical Std: {theoretical_std:.4f}")
print(f"Sample Std: {sample_std:.4f}")

# Perform a Kolmogorov-Smirnov test against a uniform distribution
ks_statistic, p_value = stats.kstest(data, 'uniform', args=(low, high-low)) # :cite[3]
print(f"\nKolmogorov-Smirnov Test:")
print(f"P-value: {p_value:.4f}")
# A high p-value (typically > 0.05) suggests the data follows the specified distribution.
if p_value > 0.05:
    print("The data is consistent with a uniform distribution.")
else:
    print("The data is not consistent with a uniform distribution.")
````

###  Characteristics of the Uniform Distribution

- **Equally Likely Outcomes**: Every value within the defined range `[a, b]` has an equal probability of occurring .
- **Probability Density Function (PDF)**: The PDF is constant and given by `f(x) = 1 / (b - a)` for `a  x  b` .
- **Shape**: The graph of the PDF is a rectangle, giving it the name "rectangular distribution" .
- **Parameters**: It is defined by two parameters: the lower bound `a` and the upper bound `b` .
- **Mean and Variance**: The mean is `(a + b) / 2`, and the variance is `(b - a) / 12` .

###  Common Use Cases

The uniform distribution is widely used in various fields for modeling and simulation.

| Use Case Category | Examples |
| :--- | :--- |
| **Random Selection** | Rolling a fair die, drawing a card from a well-shuffled deck, or simple lottery systems . |
| **Computer Science** | Random number generators and load balancing algorithms where equal probability is required . |
| **Simulation & Modeling** | Monte Carlo simulations to model processes where no prior information favors any particular outcome . |

### __2. Right Skewed__

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate data from a perfect right-skewed distribution (Exponential)
np.random.seed(42)  # Ensures reproducible results
scale = 2.0         # Scale parameter (mean = 1/scale)
sample_size = 1000
data = np.random.exponential(scale=scale, size=sample_size)

# 2. Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.7, 
         edgecolor='black', color='coral')
plt.title('Histogram of a Right-Skewed Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)
plt.show()

# 3. Simple Statistical Verification
# Calculate skewness and basic statistics
sample_skewness = stats.skew(data)
sample_mean = np.mean(data)
sample_median = np.median(data)

print("Statistical Verification Results:")
print(f"Sample Skewness: {sample_skewness:.4f}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Median: {sample_median:.4f}")

# Check if distribution is right-skewed
if sample_skewness > 0:
    print(" Distribution is right-skewed (positive skewness)")
if sample_mean > sample_median:
    print(" Mean > Median (characteristic of right-skewed distributions)")

# KS test against exponential distribution
ks_statistic, p_value = stats.kstest(data, 'expon', args=(0, scale))
print(f"\nKolmogorov-Smirnov Test P-value: {p_value:.4f}")
if p_value > 0.05:
    print(" The data is consistent with an exponential distribution.")
else:
    print(" The data does not follow an exponential distribution.")
````

## Characteristics of Right-Skewed Distributions

- **Tail Direction**: Long tail extends to the right (positive side)
- **Mean > Median > Mode**: Mean is pulled in the direction of the tail
- **Positive Skewness**: Skewness coefficient > 0
- **Common Shapes**: Exponential, log-normal, gamma distributions
- **Data Concentration**: Most data points cluster on the left side

## Common Use Cases

| Category | Examples |
|----------|----------|
| **Income & Wealth** | Household incomes, wealth distribution |
| **Insurance** | Claim amounts, disaster losses |
| **Web Analytics** | Page view durations, user session times |
| **Biology** | Species abundance, lifespan distributions |
| **Finance** | Stock returns, loan amounts |
| **Social Media** | Post engagement, follower counts |

### __2. Right Skewed Distribution__

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate data from a perfect left-skewed distribution (Beta distribution)
np.random.seed(42)  # Ensures reproducible results
alpha = 8.0         # Shape parameter alpha
beta = 2.0          # Shape parameter beta
sample_size = 1000
data = np.random.beta(alpha, beta, size=sample_size)

# 2. Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.7, 
         edgecolor='black', color='lightblue')
plt.title('Histogram of a Left-Skewed Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)
plt.show()

# 3. Simple Statistical Verification
# Calculate skewness and basic statistics
sample_skewness = stats.skew(data)
sample_mean = np.mean(data)
sample_median = np.median(data)

print("Statistical Verification Results:")
print(f"Sample Skewness: {sample_skewness:.4f}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Median: {sample_median:.4f}")

# Check if distribution is left-skewed
if sample_skewness < 0:
    print(" Distribution is left-skewed (negative skewness)")
if sample_mean < sample_median:
    print(" Mean < Median (characteristic of left-skewed distributions)")

# KS test against beta distribution
ks_statistic, p_value = stats.kstest(data, 'beta', args=(alpha, beta))
print(f"\nKolmogorov-Smirnov Test P-value: {p_value:.4f}")
if p_value > 0.05:
    print(" The data is consistent with a beta distribution.")
else:
    print(" The data does not follow a beta distribution.")
````

## Characteristics of Left-Skewed Distributions

- **Tail Direction**: Long tail extends to the left (negative side)
- **Mode > Median > Mean**: Mean is pulled in the direction of the tail
- **Negative Skewness**: Skewness coefficient < 0
- **Data Concentration**: Most data points cluster on the right side
- **Also Called**: Negative skew distribution

## Common Use Cases

### __2. Bimodal Distribution__

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate data from a perfect bimodal distribution (two normal distributions combined)
np.random.seed(42)  # Ensures reproducible results
sample_size = 1000

# Create two different normal distributions and combine them
data1 = np.random.normal(loc=10, scale=1.5, size=sample_size//2)
data2 = np.random.normal(loc=20, scale=1.5, size=sample_size//2)
data = np.concatenate([data1, data2])

# 2. Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.7, 
         edgecolor='black', color='lightgreen')
plt.title('Histogram of a Bimodal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)
plt.show()

# 3. Simple Statistical Verification
# Calculate basic statistics and check for multiple peaks
sample_mean = np.mean(data)
sample_median = np.median(data)

# Use kernel density estimation to detect peaks
kde = stats.gaussian_kde(data)
x_vals = np.linspace(data.min(), data.max(), 1000)
kde_vals = kde(x_vals)

# Find peaks in the KDE (simplified approach)
from scipy.signal import find_peaks
peaks, _ = find_peaks(kde_vals, height=0.05)
num_peaks = len(peaks)

print("Statistical Verification Results:")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Median: {sample_median:.4f}")
print(f"Number of detected peaks: {num_peaks}")

if num_peaks >= 2:
    print(" Distribution is multimodal (bimodal or more)")
    print(" Two distinct peaks visible in the histogram")
else:
    print(" Distribution does not show clear multimodality")
````

## Characteristics of Bimodal Distributions

- **Multiple Peaks**: Two or more distinct peaks in the distribution
- **Mixed Populations**: Represents combination of two or more different groups
- **Valley Between Modes**: Trough between the peaks indicates separation
- **Common Shapes**: Often appears as two humps in the histogram
- **Statistical Significance**: Indicates underlying heterogeneity in data

# The Importance of Normal Distribution in Data Science

##  What is Normal Distribution?

The **Normal Distribution** (also called Gaussian distribution) is the most famous probability distribution in statistics. It's that familiar bell-shaped curve you've probably seen before!

````python
# Simple visual example of normal distribution
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.7)
plt.title('The Famous Bell Curve - Normal Distribution')
plt.show()
````

##  Why is Normal Distribution So Special?

### 1. **It's Everywhere in Nature**
Think about these everyday examples:
- **Heights** of people in a population
- **Blood pressure** readings
- **Test scores** in a large class
- **Measurement errors** in experiments

Most natural phenomena tend to cluster around an average value, with fewer cases at the extremes.

### 2. **The Central Limit Theorem - The Magic Rule**
This is the **most important reason** normal distribution matters:

> "When you take large enough random samples from ANY population, the means of those samples will form a normal distribution - regardless of what the original population looks like."

**Simple Example:** 
- Roll a die once  equal probability (uniform distribution)
- Roll 10 dice and take the average  starts looking like normal distribution
- Roll 100 dice and take the average  perfect bell curve!

### 3. **Statistical Tests Depend on It**
Most common statistical tests assume normality:
- **t-tests** (comparing two groups)
- **ANOVA** (comparing multiple groups)
- **Regression analysis** (predicting relationships)
- **Confidence intervals**

##  Practical Importance in Data Science

### **Machine Learning**
- Many algorithms assume features are normally distributed
- **Example:** Gradient Descent works better when features are normalized
- Algorithms like Gaussian Naive Bayes explicitly use normal distribution

### **Data Preprocessing**


````python
# Why we often transform data to be normal
from scipy import stats

# Check if data is normal
p_value = stats.normaltest(data).pvalue
if p_value > 0.05:
    print("Data is normally distributed - great for modeling!")
else:
    print("Might need transformation for better results")
````

### **Outlier Detection**
Normal distribution gives us the **68-95-99.7 Rule**:
- 68% of data falls within **1 standard deviation** of mean
- 95% of data falls within **2 standard deviations** of mean  
- 99.7% of data falls within **3 standard deviations** of mean

Any data point beyond 3 standard deviations is probably an outlier!

##  Real-World Data Science Applications

### **A/B Testing**
When comparing website conversion rates:
- We assume the difference between groups follows normal distribution
- This lets us calculate p-values and determine if results are significant

### **Quality Control**
Manufacturing companies use normal distribution to:
- Set acceptable quality ranges
- Detect when a process is going out of control
- Monitor product dimensions and weights

### **Risk Management**
Banks and insurance companies use it for:
- Calculating investment risks
- Estimating insurance claim amounts
- Modeling stock price movements

##  Important Caveats

### **When Data Isn't Normal**
Not everything follows normal distribution!
- **Income data**  usually right-skewed
- **Insurance claims**  often right-skewed  
- **Survival times**  can be various shapes

### **Solutions for Non-Normal Data**
1. **Transformations** (log, square root, Box-Cox)
2. **Use non-parametric tests** that don't assume normality
3. **Try different distributions** that fit your data better

##  Key Takeaways for Students

1. **Normal distribution is the default assumption** in many statistical methods
2. **Central Limit Theorem** makes it incredibly useful for working with sample means
3. **Many machine learning algorithms** perform better with normally distributed data
4. **It provides a framework** for understanding variability and detecting outliers
5. **Always check if your data is actually normal** before applying normal-based methods

##  Remember This Analogy

> **Normal distribution is like the "standard ruler" of statistics. Just like we use standard meters and kilograms for measurements, we use normal distribution as our standard reference for understanding data patterns.**

The normal distribution isn't just a mathematical concept - it's the fundamental language that helps data scientists communicate with data, build reliable models, and make confident decisions!

# Data Normalization: Making Your Data Speak the Same Language

##  What is Data Normalization?

Imagine you're comparing students' performance, but one exam is out of **100 points** and another is out of **20 points**. It's confusing, right? **Normalization** solves this problem by putting all your data on the same scale!


````python
# Before Normalization - Different scales, hard to compare
math_scores = [85, 92, 78, 96]     # Scale: 0-100
quiz_scores = [17, 19, 15, 18]     # Scale: 0-20

# After Normalization - Same scale, easy to compare!
normalized_math = [0.7, 0.9, 0.6, 1.0]    # All between 0-1
normalized_quiz = [0.7, 0.9, 0.6, 1.0]    # All between 0-1
````

##  Why Do We Normalize Data?

### **1. Fair Comparison**
Like comparing apples to apples, not apples to elephants!

### **2. Machine Learning Loves Consistency**
Most algorithms work better when features are on similar scales:
- **Distance-based algorithms** (KNN, K-means)
- **Gradient Descent** converges faster
- **Regularization** works properly

### **3. Prevents Feature Domination**
Without normalization, features with larger ranges can unfairly dominate:
- **Salary** (30,000-100,000) would dominate over **Age** (20-60)
- The algorithm might think salary is 1000x more important!

##  Common Normalization Techniques

### **1. Min-Max Scaling (0 to 1 Range)**
**Formula:** `(x - min) / (max - min)`


````python
# Simple Min-Max Scaling Example
def min_max_scale(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

# Example: Test scores from different classes
class_A = [55, 60, 65, 70, 85]    # Hard exam
class_B = [80, 85, 90, 95, 98]    # Easy exam

normalized_A = min_max_scale(class_A)  # [0.0, 0.17, 0.33, 0.5, 1.0]
normalized_B = min_max_scale(class_B)  # [0.0, 0.28, 0.56, 0.83, 1.0]
````

**Best for:** When you know your boundaries (like percentages 0-100%)

### **2. Standardization (Z-score Normalization)**
**Formula:** `(x - mean) / standard deviation`

````python
# Standardization puts data around mean=0, std=1
def standardize(data):
    mean = sum(data) / len(data)
    std = (sum((x - mean)**2 for x in data) / len(data)) ** 0.5
    return [(x - mean) / std for x in data]

# Example: Student heights in different classes
grade_5_heights = [130, 135, 140, 145, 150]  # cm
grade_10_heights = [160, 165, 170, 175, 180] # cm

standardized_5 = standardize(grade_5_heights)   # Mean  0, Std  1
standardized_10 = standardize(grade_10_heights) # Mean  0, Std  1
````

**Best for:** When data has outliers or unknown boundaries

### **3. Robust Scaling**
Uses median and interquartile range - great when you have outliers!

````python
# Robust scaling handles outliers better
def robust_scale(data):
    from numpy import percentile
    Q1 = percentile(data, 25)
    Q3 = percentile(data, 75)
    IQR = Q3 - Q1
    median = percentile(data, 50)
    return [(x - median) / IQR for x in data]
````

# Example: Income data with billionaire CEO
salaries = [50000, 55000, 60000, 65000, 1000000000]
normalized = robust_scale(salaries)  # Outlier has less extreme effect


##  When to Use Which Method?

| Method | Best For | Watch Out For |
|--------|----------|---------------|
| **Min-Max** | Images (0-255), Percentages, Known bounds | Outliers can ruin the scaling |
| **Standardization** | Most machine learning, Normal distributions | Doesn't guarantee specific range |
| **Robust Scaling** | Data with outliers, Non-normal data | Less common, harder to explain |

##  Real Data Science Example


````python
# Before normalization - machine learning struggle
features_before = {
    'age': [25, 30, 35, 40],           # Range: 25-40
    'salary': [50000, 80000, 60000, 120000],  # Range: 50,000-120,000
    'credit_score': [650, 700, 720, 680]      # Range: 650-720
}

# After normalization - happy machine learning!
features_after = {
    'age': [0.0, 0.33, 0.67, 1.0],
    'salary': [0.0, 0.43, 0.14, 1.0],
    'credit_score': [0.0, 0.71, 1.0, 0.43]
}
````

# Symmetrical Data Distribution

##  Simple Definition

**Symmetrical Distribution** means your data is perfectly balanced around the center. If you fold it in half at the middle, both sides match exactly like a mirror image.

##  Standard Definition

A probability distribution where the values are equally distributed on both sides of the central point (mean), creating a mirror image when split down the center.

##  Easy Explanation

Think of a seesaw that's perfectly balanced:
- **Left side** = **Right side** (equal weight)
- **Center point** doesn't tilt in either direction
- **Mean, Median, and Mode** all fall at the same center point

```
Left Side:  [5, 6, 7, 8, 9]
Center:     10
Right Side: [11, 12, 13, 14, 15]
```

##  Visual Characteristics

- **Bell-shaped curve** (most common)
- **Equal tails** on both ends
- **Balanced spread** of data points
- **Center line** divides data equally

##  Key Properties

1. **Mean = Median = Mode** (all center measures match)
2. **Zero Skewness** (no leaning to left or right)
3. **Mirror Image** on both sides of center
4. **Equal Probability** of extreme values in both directions

##  Common Examples

### **Nature & Biology**
- **Human Height** in a large population
- **Blood Pressure** readings in healthy adults
- **Leaf Sizes** on the same tree species
- **Animal Weights** within same species

### **Education**
- **Well-designed Test Scores** (when exam is neither too hard nor too easy)
- **IQ Scores** in general population
- **Class Attendance** patterns

### **Manufacturing & Quality**
- **Precision Parts** measurements (screw lengths, bolt diameters)
- **Product Weights** in quality control
- **Assembly Times** for standardized tasks

### **Everyday Life**
- **Commute Times** on predictable routes
- **Coffee Temperatures** from same machine
- **Pizza Delivery Times** in normal conditions

##  The Balance Test

**Your data is symmetrical if:**
-  Histogram looks the same on left and right
-  Mean and median are very close
-  No long tail on either side
-  Equal number of extreme values in both directions

**Simple Check:**
```
If Mean  Median  Symmetrical
If Mean > Median  Right-skewed (not symmetrical)
If Mean < Median  Left-skewed (not symmetrical)
```

__Symmetrical distribution is the "gold standard" in statistics because it makes analysis simpler and predictions more reliable!__

# Asymmetrical Data Distribution

##  Simple Definition

**Asymmetrical Distribution** means your data is unbalanced and leans to one side. It's like a seesaw with more weight on one end - the data doesn't form a mirror image around the center.

##  Standard Definition

A probability distribution where the values are not equally distributed on both sides of the central tendency, resulting in skewness and lack of mirror symmetry around the mean.

##  Easy Explanation

Think of a lopsided seesaw:
- **One side is heavier** (more data points)
- **One side is lighter** (fewer data points)
- The **balance point shifts** toward the heavier side
- The curve "leans" or "tails off" in one direction

##  Visual Characteristics

- **Uneven tails** - one side stretches longer
- **Peak shifted** away from center
- **No mirror image** when split down middle
- **Mean  Median  Mode** (they separate)

##  Types of Asymmetry

### **Right-Skewed (Positive Skew)**
```
     /
    /
___/
```
- **Tail extends to RIGHT**
- **Mean > Median > Mode**
- **Most data clustered on LEFT**

### **Left-Skewed (Negative Skew)**
```
\__
           \
            \
```
- **Tail extends to LEFT**
- **Mode > Median > Mean**
- **Most data clustered on RIGHT**

##  Common Examples

### **Right-Skewed (Positive Skew) Examples**
- **Income Distribution** (few rich, many middle/poor)
- **House Prices** (few mansions, many average homes)
- **Social Media Likes** (few viral posts, many with few likes)
- **Insurance Claims** (few large claims, many small ones)
- **Hospital Stay Costs** (few very expensive stays)

### **Left-Skewed (Negative Skew) Examples**
- **Age at Death** in developed countries (most live long)
- **Exam Scores** when test is too easy (most score high)
- **Retirement Age** (most retire around same age)
- **Product Failure Times** (most fail late in life)
- **Customer Ratings** for popular products (mostly 4-5 stars)

##  Key Properties

1. **Skewness  0** (measures the asymmetry)
2. **Mean, Median, Mode separate**
3. **Tail indicates direction of skew**
4. **Natural limits often cause asymmetry** (can't have negative income)

##  The Imbalance Test

**Your data is asymmetrical if:**
-  Histogram looks lopsided
-  Mean and median are far apart
-  One tail is much longer
-  Most data piles up on one side

**Quick Direction Check:**
```
If Mean > Median  Right-skewed
If Mean < Median  Left-skewed
```

##  Why It Matters

- **Statistical methods** may need adjustment
- **Transformations** often needed (log, square root)
- **Different analysis approaches** required
- **Real-world data** is often asymmetrical

Asymmetrical distributions are actually more common in real-world data than perfect symmetrical ones! Understanding the "lean" of your data helps choose the right analytical approach.

# __Skewness VS Kurtosis__

## __1. Skewness: The Asymmetry Measure__

- **Definition**: Skewness measures the degree of asymmetry of a distribution around its mean. It indicates whether the data points are skewed to the left (negative skew) or to the right (positive skew) of the mean.

- **Positive Skew**: A distribution with a longer tail on the right side.

- **Negative Skew**: A distribution with a longer tail on the left side.

- **Example**: Income distribution is often positively skewed, as a majority of people earn below the average, with a few high earners creating a long right tail.

# Skewness Formula

The skewness of a dataset can be calculated using the following formula:

$ Skewness(G) = \frac{n}{(n-1)(n-2)} \sum_{i=1}^n \left(\frac{x_i - \bar{x}}{s}\right)^3 $

Where:

- $ ( n ) $ is the number of data points in the dataset.  
- $ ( x_i ) $ represents each individual data point.  
- $ ( \bar{x} ) $ is the mean of the dataset.  
- $ ( s ) $ is the standard deviation of the dataset.

This formula calculates the standardized third moment (the sum of the cubed deviations from the mean, divided by the standard deviation cubed), adjusted for bias in small samples.

# 2. Kurtosis: The Tailedness Measure

- **Definition**: Kurtosis measures the 'tailedness' of a distribution. It describes the height and sharpness of the central peak and the heaviness of the distribution's tails.

- **High Kurtosis (Leptokurtic)**: A distribution with heavy tails and a sharper peak than a normal distribution.

- **Low Kurtosis (Platykurtic)**: A distribution with lighter tails and a flatter peak.

- **Mesokurtic**: A distribution with kurtosis similar to a normal distribution.

- **Example**: A dataset of exam scores where most students performed similarly (high Kurtosis) versus a dataset with more varied scores (low Kurtosis).

# Kurtosis Formula

Kurtosis is calculated using this formula:

$ Kurtosis \left( K \right) = \frac{n(n+1)}{(n-1)(n-2)(n-3)} \sum_{i=1}^{n} \left( \frac{x_i - \bar{x}}{s} \right)^4 - \frac{3(n-1)^2}{(n-2)(n-3)} $

Where:

- $ ( n ) $ is the number of data points in the dataset  
- $ ( x_i ) $ represents each individual data point  
- $ ( \bar{x} ) $ is the mean of the dataset  
- $ ( s ) $ is the standard deviation of the dataset

This formula calculates the standardized fourth moment (the sum of the fourth power of deviations from the mean, divided by the standard deviation to the fourth power). The term at the end adjusts for the kurtosis of a normal distribution, making the kurtosis of the normal distribution zero (this is sometimes called "excess kurtosis").

# The Role of Skewness and Kurtosis in Data Analysis

- **Understanding Data Distribution**: Skewness and kurtosis provide insight into the data's distribution pattern, which is crucial for selecting the right statistical tests and models.

- **Identifying Outliers**: Skewness, in particular, can help in detecting outliers and understanding their impact on the dataset.

- **Predictive Modelling**: In machine learning and predictive modeling, knowing the skewness and kurtosis can guide data transformation and normalization processes.

````python
import numpy as np
from scipy.stats import skew, kurtosis

# Generating example data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000) # Normal distribution (kurtosis ~ 3)
 
# Caculating skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data, fisher=True) # Fisher=True gives excess kurtosis (normal = 0)

# Displaying results
print(f"Skewness: {data_skewness:.4f}")
print(f"Kurtosis (Excess): {data_kurtosis:.4f}")
````

````python
sns.histplot(data, kde=True)
````

# How to Test the Normal Distribution of Data
- ### __H0: Data is normally distributed.__
- ### __H1: Data is not normally distributed.__
- ### __Shapiro-Wilk Test:__
- ### __d'Agostino's K-squared Test:__
- ### __Anderson-Darling Test:__

````python
# Shapiro-Wilk Test to test the normality of data
stats.shapiro(data)
print(f"Shapiro-wilk test statistic: {stats.shapiro(data)[0]}")
print(f"Shapiro-wilk P-value: {stats.shapiro(data)[1]}")
````

Explanation of Tests:
- __Shapiro-Wilk Test:__ This test assesses whether a sample comes from a normally distributed population. It is particularly effective for small sample sizes (n < 50). The null hypothesis (H0) states that the data is normally distributed. A p-value less than the chosen alpha level (commonly 0.05) leads to the rejection of H0, indicating that the data is not normally distributed.
- __d'Agostino's K-squared Test:__ This test evaluates the skewness and kurtosis of the data to determine if it deviates from a normal distribution. It is suitable for larger sample sizes (n > 20). The null hypothesis (H0) states that the data is normally distributed. A p-value less than the alpha level results in the rejection of H0, suggesting non-normality.
- __Anderson-Darling Test:__ This test compares the empirical distribution of the data to a specified theoretical distribution (often normal). It provides a test statistic and critical values for various significance levels. If the test statistic exceeds the critical value for a chosen significance level, H0 is rejected, indicating that the data does not follow the specified distribution.
