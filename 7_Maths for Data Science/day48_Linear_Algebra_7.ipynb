## 1. What is a System of Equations?

### Definition
A system of equations is like a puzzle where we have two or more equations with the same variables, and we need to find values for those variables that make ALL equations true.

**Simple Example:**
- Equation 1: `x + y = 5`
- Equation 2: `x - y = 1`

We need to find numbers for x and y that work in both equations.

### Visual Example
Let's use Python to show what this looks like:

````python
import matplotlib.pyplot as plt
import numpy as np

# Create x values
x = np.linspace(0, 5, 100)

# Create y values for each equation
y1 = 5 - x  # from x + y = 5
y2 = x - 1  # from x - y = 1

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='x + y = 5', linewidth=2, color='blue')
plt.plot(x, y2, label='x - y = 1', linewidth=2, color='red')

# Mark the solution point
plt.scatter(3, 2, color='green', s=100, zorder=5, label='Solution (3,2)')

# Add labels and grid
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Solving Systems of Equations Graphically')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Set the axis limits
plt.xlim(0, 5)
plt.ylim(0, 5)

# Show the plot
plt.show()

# This code creates a graph showing our two equations. The green dot where they cross is our solution!
````

## 1. Graphical Method

### Definition
Solving by drawing the equations on a graph and finding where they cross.

### How It Works
1. Turn each equation into `y = ` form
2. Draw each line on graph paper
3. Find where they cross - that's your answer!

### Example
Let's solve:
- `x + y = 5`  `y = 5 - x`
- `x - y = 1`  `y = x - 1`

We draw both lines and see they cross at (3,2). So `x = 3` and `y = 2`.

### Benefits
- Easy to understand
- You can see the answer
- Helps visualize what's happening

### Limitations
- Not exact if lines cross between whole numbers
- Hard with more than 2 variables
- Can be time-consuming to draw

### When to Use
- When you want to visualize the problem
- For simple 2-variable systems
- When you need a rough estimate

### Tips
- Use graph paper for accuracy
- Check your answer by plugging it into both equations
- Use different colors for different lines

## 2. Substitution Method

### Definition
Solving by getting one variable alone and "substituting" it into the other equation.

### How It Works
1. Solve one equation for one variable
2. Replace that variable in the other equation
3. Solve for the remaining variable
4. Find the other variable

### Example
Let's solve:
- `x + y = 5`
- `x - y = 1`

**Step 1:** Solve first equation for x: `x = 5 - y`

**Step 2:** Substitute into second equation: `(5 - y) - y = 1`

**Step 3:** Solve: `5 - 2y = 1`  `-2y = -4`  `y = 2`

**Step 4:** Find x: `x = 5 - 2`  `x = 3`

**Check:** `3 + 2 = 5`  and `3 - 2 = 1` 

### Benefits
- Always works (if there's a solution)
- Exact answers
- Good when one variable is easy to isolate

### Limitations
- Can get messy with fractions
- Tricky with more than 2 variables
- Might need to work with decimals

### When to Use
- When one variable is easy to isolate
- For exact solutions
- When equations are already in a convenient form

### Tips
- Choose the equation that's easiest to solve first
- Always check your answer in both original equations
- If you get fractions, that's okay!

---

## 3. Elimination Method

### Definition
Solving by adding or subtracting equations to "eliminate" one variable.

### How It Works
1. Line up equations with like terms
2. Multiply if needed to make coefficients match
3. Add or subtract to eliminate one variable
4. Solve for the remaining variable
5. Find the other variable

### Example
Let's solve:
- `x + y = 5`
- `x - y = 1`

**Step 1:** Already lined up!

**Step 2:** Add the equations:
```
 x + y = 5
+x - y = 1
---------
2x     = 6
```

**Step 3:** Solve: `2x = 6`  `x = 3`

**Step 4:** Find y: `3 + y = 5`  `y = 2`

**Check:** `3 + 2 = 5`  and `3 - 2 = 1` 

### Benefits
- Works well when coefficients match
- Good for avoiding fractions
- Often the fastest method

### Limitations
- Sometimes needs multiplication
- Can be tricky with more variables
- Might create larger numbers

### When to Use
- When coefficients are similar
- When you want to avoid fractions
- When equations are already lined up nicely

### Tips
- Look for easiest variable to eliminate first
- Multiply by smallest number possible
- Always check your work

## 4. Matrix Method (Inverse Method)

### Definition
Using a special matrix called the "inverse" to solve equations, like using a master key to open a lock.

### How It Works
1. Write equations as: **A**  **X** = **B**
2. Find the inverse of **A** (called **A**)
3. Multiply both sides by **A**
4. Get: **X** = **A**  **B**

### Example
Let's solve:
- `2x + y = 5`
- `x - y = 1`

**Step 1:** Write in matrix form:
```
[2  1] [x] = [5]
[1 -1] [y]   [1]
```

**Step 2:** Find inverse of `[2  1]`
                        `[1 -1]`

Formula for 2x2 inverse: 
```
[a  b] = 1/(ad-bc)  [d -b]
[c  d]                [-c a]
```

So: `1/((2)(-1) - (1)(1)) = 1/(-3) = -1/3`

And: `[-1  -1]  -1/3 = [1/3  1/3]`
     `[-1   2]         [1/3 -2/3]`

**Step 3:** Multiply inverse by `[5]`
                             `[1]`

```
[1/3  1/3] [5] = [5/3 + 1/3] = [6/3] = [2]
[1/3 -2/3] [1]   [5/3 - 2/3]   [3/3]   [1]
```

**Answer:** x = 2, y = 1

### Benefits
- Works like magic once you know the inverse
- Very neat and organized
- Great for computers

### Limitations
- Only works for square matrices
- Can be tricky to find inverses
- Not all matrices have inverses

### When to Use
- When you have the same number of equations and variables
- When working with calculators or computers

### Tips
- Practice with 2x2 matrices first
- Always check if the determinant (ad-bc) is not zero
- Use this method when you can use a calculator
