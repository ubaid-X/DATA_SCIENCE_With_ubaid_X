> # Zero-Math

📘 What is Mathematics?

Mathematics is the science of numbers, shapes, patterns, and logic.  
It helps us count, measure, analyze, and solve problems in real life and science.

---

🧩 Key Factors of Mathematics:

---

1. 🔢 Counting (Arithmetic / Numbers)

Used to count objects or people.

- Example: 1, 2, 3, 4, …  
- Use Case: Counting apples: 🍎🍎🍎 → 3 apples

---

2. 📏 Measurement

Used to find the length, weight, time, area, volume, etc.

- Example:  
  - Length = 5 cm  
  - Time = 2 hours  
  - Volume = 1 liter  
- Use Case: Measuring the height of a table → 90 cm

---

3. ➗ Quantity

Deals with how much of something.

- Example:  
  - 4 liters of milk  
  - 10 rupees  
- Use Case: Comparing 5 mangoes 🍊 with 3 oranges 🍊 → 5 > 3

---

4. 🧠 Logical Reasoning

Thinking step by step to reach a correct conclusion.

- Example:  
  - If A > B and B > C → then A > C  
- Use Case: Solving puzzles or math word problems

---

5. 🧱 Structure (Algebra)

Understanding how numbers and variables relate in equations or formulas.

- Example:
   - Equation: x + 5 = 10 → x = 5  
- Use Case: Finding unknown values in business or science

---

7. 🧪 Proof (Rigorous Thinking)

Proving something is always true using logical steps.

- Example:  
  - Sum of angles in triangle = 180°  
- Use Case: In geometry to prove shapes' properties

---

7. 🔁 Patterns and Sequences

Recognizing how numbers or shapes repeat or grow.

- Example:  
  - 2, 4, 6, 8, … (pattern: +2)  
- Use Case: Predicting next value in data or designs

---

8. 🧮 Problem Solving

Using math to solve real-world problems.

- Example:  
  - You have 100 Rs, buy 3 pencils  15 Rs each → how much left?  
  - 3 × 15 = 45 → 100 - 45 = 55 Rs

---

📘 Number Theory — Definition & Types

Number theory is a branch of mathematics that deals with properties of numbers, especially integers (like 1, 2, 3, …).  
It helps us classify numbers, study their patterns, and explore how they behave.

---

🔢 1. Natural Numbers (ℕ)

- Definition: Counting numbers starting from 1.
- Example: 1, 2, 3, 4, 5, …

✅ Used for: Counting apples, people, books etc

---

🧱 2. Whole Numbers

- Definition: Natural numbers + Zero
- Example: 0, 1, 2, 3, 4, …

✅ Used for: Number of pages (can be 0)

---

🔁 3. Even Numbers

- Definition: Numbers divisible by 2  
- Symbolically: n % 2 == 0
- Example: 2, 4, 6, 8, 10, …

✅ Used in: Arranging things in pairs

---

➖ 4. Odd Numbers

- Definition: Numbers that leave remainder 1 when divided by 2  
- Symbolically: n % 2 == 1
- Example: 1, 3, 5, 7, 9, …

---

🔢 5. Prime Numbers

- Definition: Numbers with exactly 2 factors (1 and itself)  
- Example: 2, 3, 5, 7, 11, 13, …

✅ Note: 2 is the only even prime number

---

🔁 6. Composite Numbers

- Definition: Numbers with more than 2 factors  
- Example: 4 → (1, 2, 4), 6 → (1, 2, 3, 6)

❌ Not Prime

---

🔢 7. Integers (ℤ)  

Definition: Numbers without decimals or fractions. Includes negative, zero, and positive values.  

Symbol: ℤ  

Examples: -5, -2, 0, 3, 100  

Use Cases:  
- Counting money (profit/loss)  
- Temperature differences  
- Time zones  

---

➗ 8. Rational Numbers (ℚ)  

Definition: Numbers that can be expressed as a fraction (p/q), where p, q ∈ ℤ and q ≠ 0.

Symbol: ℚ  

Examples: 3/4, -2/5, 0.75 (because 3/4), 5 (because 5/1)  

Use Cases:  
- Percentages  
- Currency (₹1.25 = 5/4)  
- Ratios in recipes  

---

📏 9. Fractions  

Definition: A form of rational number that shows part of a whole, written as numerator/denominator. 

Examples: 2/3, -7/8, 5/1  

Use Cases:  
- Dividing pizza slices  
- Construction measurements  
- Time (1/4 hour = 15 minutes)  

---

🔁 10. Irrational Numbers  

Definition: Numbers that cannot be written as a fraction; their decimals never repeat or terminate.  

Examples: π ≈ 3.14159…, √2 ≈ 1.414…, e ≈ 2.718…  

Use Cases:  
- Geometry (diagonal of square = √2)  
- Trigonometry (angles with π)
- Natural growth (e in calculus)

---

📊 11. Real Numbers (ℝ)  

Definition: All numbers that can be located on the number line. This includes both rational and irrational numbers.  

Symbol: ℝ  

Examples: -7, 0, 1/2, √5, π  

Use Cases:  
- Measuring distances, weights  
- Statistical data  
- Real-world quantities in physics/finance  

---

⚡ 12. Complex Numbers (ℂ)  

Definition: Numbers that have a real part and an imaginary part, written as a + bi where i = √-1.  

Symbol: ℂ  

Examples: 3 + 2i, -1 - 4i, i, 5  

Use Cases:  
- Electrical circuits (impedance)  
- Quantum mechanics  
- Signal processing  

---

🧮 13. Algebraic Numbers  

Definition: A number that is a root of a polynomial with integer coefficients.  

Examples:  
- √2 → root of x² - 2 = 0  
- 1/2 → root of 2x - 1 = 0  
- 
Use Cases:  
- Solving equations  
- Algebraic geometry  
- Symbolic math (CAS systems)

---

🔷 14. Square Numbers (n²)

- Definition: A number × itself  
- Example: 1² = 1, 2² = 4, 3² = 9, 4² = 16

✅ Used in: Area of square

---

🔷 15. Cube Numbers (n³)

- Definition: A number × itself × itself  
- Example: 2³ = 8, 3³ = 27, 4³ = 64

✅ Used in: Volume of cubes

---

🔄 16. Numbers ≡ 1 mod 4 and 3 mod 4

- Definition:  
  - 1 mod 4: Remainder is 1 → e.g. 5, 9, 13  
  - 3 mod 4: Remainder is 3 → e.g. 3, 7, 11  
- Meaning: n % 4 == 1 or n % 4 == 3

✅ Used in: Cryptography & primes

---

🔺 17. Triangular Numbers

- Definition: Count objects arranged in triangle  
- Formula: n(n+1)/2  
- Example: 1, 3, 6, 10, 15, …

✅ 3rd triangular number = 3(4)/2 = 6

---

💎 18. Perfect Numbers

- Definition: Sum of its proper divisors = the number  
- Example:  
  - 6 → 1 + 2 + 3 = 6  
  - 28 → 1 + 2 + 4 + 7 + 14 = 28

✅ Rare & special

---

🌀 19. Fibonacci Numbers

- Definition: Each number is sum of previous two  
- Formula: F(n) = F(n−1) + F(n−2)  
- Example: 0, 1, 1, 2, 3, 5, 8, 13, …

✅ Used in: Nature patterns (flowers, shells)
