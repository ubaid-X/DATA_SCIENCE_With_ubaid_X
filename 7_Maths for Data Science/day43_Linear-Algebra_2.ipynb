> ### __1. Cartesian Coordinates in Linear Algebra__

__Definition:__
The *Cartesian coordinate system* represents points in space using ordered pairs in 2D (or triplets in 3D). Each point is defined by its distances from perpendicular axes.

__Explanation:__
- In 2D: A point is written as *(x, y)* where:
  - *x* is the horizontal distance (along the x-axis)
  - *y* is the vertical distance (along the y-axis)
- The origin is *(0, 0)* — where the two axes meet.
- The coordinate plane is divided into *four quadrants*.

__✅ Example:__

- Point A = (3, 2): 3 units right, 2 units up
- Point B = (−4, −1): 4 units left, 1 unit down

````python
import matplotlib.pyplot as plt

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.scatter([3, -4], [2, -1], color='red')
plt.text(3, 2, 'A(3,2)', fontsize=12)
plt.text(-4, -1, 'B(-4,-1)', fontsize=12)

plt.xlim(-6, 6)
plt.ylim(-4, 4)
plt.grid(True)
plt.title('Cartesian Coordinates')
plt.show()
````

 ### __💡 Tips & Tricks__
- Quadrants:
  - I: (+, +)
  - II: (−, +)
  - III: (−, −)
  - IV: (+, −)
- Origin is always *(0, 0)*

>  ### 2. Unit Vectors

- __Definition:__
A *unit vector* is a vector with *magnitude = 1*, used to indicate direction only.

- __Explanation:__
To create a unit vector *û* from any vector *v*, divide it by its magnitude:

```math
û = V / |V|
```

__Where:__
- *|v|* is the magnitude of the vector

✅ Example
Given *v = (3, 4)*:

```math
|v⃗| = √(3^2 + 4^2) = √(25) = 5
```

```math
û = (3/5, 4/5)
```

````python
import numpy as np

v = np.array([3, 4])
unit_v = v / np.linalg.norm(v)

plt.quiver(0, 0, *v, color='blue', scale=1, angles='xy', scale_units='xy', label='v = (3,4)')
plt.quiver(0, 0, *unit_v, color='green', scale=1, angles='xy', scale_units='xy', label='Unit Vector')

plt.xlim(0, 4)
plt.ylim(0, 5)
plt.grid(True)
plt.legend()
plt.title('Vector vs Unit Vector')
plt.show()
````

### __💡 Tips & Tricks:__
- A unit vector *only shows direction*, not magnitude.
- In 2D:
*î = (1, 0)* — x-direction
- *ĵ = (0, 1)* — y-direction

> ### __3. Scalars and Scaling a Vector__

__Definition:__
A *scalar* is a single number representing *magnitude only*.  
*Scaling a vector* means multiplying it by a scalar.

__Explanation:__

__If:__

```math
v⃗ = (x, y)
```

Then scaling by a scalar *k*:

```math
k \cdot v⃗ = (k \cdot x, k \cdot y)
```

- If *k > 1*: vector becomes longer
- If *0 < k < 1*: vector shrinks
- If *k < 0*: vector reverses direction

__Example:__

```math
v⃗ = (2, 1),\quad k = 3
```

```math
3 \cdot v⃗ = (6, 3)
```


````python
v = np.array([2, 1])
scaled_v = 3 * v

plt.quiver(0, 0, *v, color='blue', angles='xy', scale_units='xy', scale=1, label='v')
plt.quiver(0, 0, *scaled_v, color='orange', angles='xy', scale_units='xy', scale=1, label='3v')

plt.xlim(0, 7)
plt.ylim(0, 4)
plt.grid()
plt.legend()
plt.title('Vector Scaling')
plt.show()
````

__💡 Tips & Tricks__
- Scalars can stretch or shrink a vector.
- Negative scalar = vector flips direction.
- Scalar = 0 gives the *zero vector*.
