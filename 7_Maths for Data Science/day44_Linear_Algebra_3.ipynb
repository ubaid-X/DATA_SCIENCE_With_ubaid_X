> ### __1. Vector Addition__

__Definition:__
Vector addition is the process of combining two vectors to form a resultant vector.

__Explanation:__

If *v = (v‚ÇÅ, v‚ÇÇ)* and *u = (u‚ÇÅ, u‚ÇÇ)*, then:

```math
v‚Éó + u‚Éó = (v_1 + u_1,\ v_2 + u_2)
```

__Example:__

```math
v‚Éó = (3, 2),\quad u‚Éó = (1, 4)
```

```math
v‚Éó + u‚Éó = (3+1,\ 2+4) = (4, 6)
```

````python
import matplotlib.pyplot as plt
import numpy as np

v = np.array([3, 2])
u = np.array([1, 4])
result = v + u

plt.quiver(0, 0, *v, color='blue', angles='xy', scale_units='xy', scale=1, label='v')
plt.quiver(*v, *u, color='green', angles='xy', scale_units='xy', scale=1, label='u (from v)')
plt.quiver(0, 0, *result, color='red', angles='xy', scale_units='xy', scale=1, label='v + u')

plt.xlim(0, 6)
plt.ylim(0, 7)
plt.grid()
plt.legend()
plt.title('Vector Addition')
plt.show()
````

__üí° Tips and Tricks__
- Think of it as moving along v, then along u.
- Order doesn‚Äôt matter: *v + u = u + v*

> ### __2. Vector Span & Linear Dependence__

__Definition:__
- *Span*: The set of all linear combinations of given vectors.
- *Linearly Dependent*: One vector can be written as a combination of others.
- *Linearly Independent*: No vector in the set can be written using the others.

__Explanation:__
Vectors *v‚ÇÅ, v‚ÇÇ, ..., v‚Çô* are linearly dependent if:

```math
c_1 v‚Éó_1 + c_2 v‚Éó_2 + \dots + c_n v‚Éó_n = 0‚Éó
```

...has a non-zero solution (not all c·µ¢ = 0).

__Example:__

```math
v‚Éó_1 = (2, 4),\quad v‚Éó_2 = (1, 2)
```
Here, *v‚ÇÅ = 2 √ó v‚ÇÇ* ‚Üí They are *linearly dependent*.

---

üí° Tips
- Use determinant or rank tests in higher dimensions.
- Independent vectors create a *basis*.

> # __3. Vector Space__

__Definition:__
A *vector space* is a collection of vectors that can be added and scaled and still remain in the space.

__Requirements (Properties):__
1. Closed under addition
2. Closed under scalar multiplication
3. Contains a zero vector
4. Obeys vector space axioms (associativity, distributivity, etc.)

__Example:__

Set of all 2D vectors *(x, y)* over real numbers *‚Ñù¬≤* is a vector space.

---

__üí° Tips and Tricks:__
- Common spaces: *‚Ñù¬≤*, *‚Ñù¬≥*, matrix space, function space
- Vectors must remain inside the set under all operations

> # __4. X and Y Intercepts in Linear Algebra__

__Definition:__
- *X-intercept*: Point where line crosses x-axis (*y = 0*)
- *Y-intercept*: Point where line crosses y-axis (*x = 0*)

__Explanation:__

Given a linear equation:

```math
y = mx + b
```

- *Y-intercept* is *b*
- To find X-intercept, set *y = 0*, solve for *x*

__Example:__

```math
y = 2x - 4
```
- Y-intercept: (0, -4)
- X-intercept:  
  Set y = 0  
  ```math
  0 = 2x - 4 \Rightarrow x = 2  
  ‚Üí X-intercept = (2, 0)
  ```

---

__üí° Tips and Tricks__
- Always set the *other variable to zero* to find an intercept.
- Intercepts help sketch linear graphs quickly.

> # __5. Dot Product of Vectors__

__Definition:__
The *dot product* of two vectors returns a *scalar*, representing projection and angle relation.

__Formula:__

```math
a‚Éó \cdot b‚Éó = a_x b_x + a_y b_y
```

Or __using angle:__

```math
a‚Éó \cdot b‚Éó = |a‚Éó| |b‚Éó| \cos(\theta)
```

__Example 1:__
```math
a‚Éó = (3, 4),\quad b‚Éó = (2, -1)
```

```math
a‚Éó \cdot b‚Éó = 3√ó2 + 4√ó(‚Äì1) = 6 ‚Äì 4 = 2
```

---

__Example 2:__
Let  
*A = (2, 3, 4)*  
*B = (1, 0, -2)*  
We want to find *A ¬∑ B*

---

‚úÖ *Step-by-Step Explanation*

*Step 1: Understand the Dot Product*

The *dot product* (also called scalar product) is a way to multiply two vectors that gives a *single number (a scalar)*.

Formula:

```math
A‚Éó \cdot B‚Éó = A_x B_x + A_y B_y + A_z B_z
```

It measures *how much one vector goes in the direction of another*.

---

*Step 2: Plug in the Components*

We are given:

- A = (2, 3, 4)  
- B = (1, 0, -2)

Now apply the formula:

```math
A \cdot B = (2 √ó 1) + (3 √ó 0) + (4 √ó -2)
```

---

*Step 3: Multiply Each Pair*

```math
= 2 + 0 - 8
```

---

*Step 4: Add the Results*

```math
= -6
```

---

‚úÖ *Final Answer:*
```math
A‚Éó \cdot B‚Éó = -6
```

---

üß† *Tips:*

- If *dot product = 0*, vectors are *perpendicular*
- Positive result ‚Üí same direction  
- Negative result ‚Üí opposite direction
- Used in projections, angles, machine learning

> # __6. Cross Product of Two Vectors__

__Definition:__
The *cross product* of two vectors results in a new vector *perpendicular* to both (only in 3D).

__Formula:__

__If:__

```math
a‚Éó = (a_1, a_2, a_3),\quad b‚Éó = (b_1, b_2, b_3)
```

__Then:__

a \times b =
\begin{vmatrix}
√Æ & ƒµ & kÃÇ \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3 \\
\end{vmatrix}

---

__Example 1:__

Let  
*A = (2, 3, 4)*  
*B = (1, 0, -2)*

We want to find *A √ó B*

---

‚úÖ *Step-by-Step Explanation*

*Step 1: Understand the Cross Product*
The *cross product* of two 3D vectors gives you another vector that is:
- *Perpendicular* to both original vectors
- Calculated using a *determinant*
- Has direction (not just magnitude)

---

*Step 2: Set up the Determinant*

We use the determinant of a 3√ó3 matrix:

```
|  i   j   k  |
|  2   3   4  |
|  1   0  -2  |
```

Where:
- `i`, `j`, `k` are the unit vectors for x, y, z directions
- The second row is vector A (2, 3, 4)
- The third row is vector B (1, 0, -2)

---

*Step 3: Expand the Determinant*

Now, apply the determinant rule:

```
A √ó B = iÃÇ (3√ó-2 ‚àí 4√ó0) ‚àí jÃÇ (2√ó-2 ‚àí 4√ó1) + kÃÇ (2√ó0 ‚àí 3√ó1)
```

Now do the math:

- iÃÇ component: (3√ó-2 ‚àí 4√ó0) = -6  
- jÃÇ component: (2√ó-2 ‚àí 4√ó1) = -4 ‚àí 4 = -8 ‚Üí minus sign makes it +8  
- kÃÇ component: (2√ó0 ‚àí 3√ó1) = -3

---

*Step 4: Write the Final Vector*

```
A √ó B = -6 iÃÇ + 8 jÃÇ ‚àí 3 kÃÇ
```

Or as a vector:

```
A √ó B = (-6, 8, -3)
```

---

üß† *Tips:*

- Use the determinant method when working in 3D.
- Always follow the *sign rules* (j term always gets a negative).
- The result vector is *perpendicular* to both input vectors.
- Cross product only works in 3D.
- The result is *orthogonal* to both vectors.
- Used in physics (torque, angular momentum).
