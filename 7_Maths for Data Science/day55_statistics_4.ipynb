### **Understanding Bias in Data**

Hello, future data scientists! Today, we're learning about **bias**. It's a small word, but it's one of the most important ideas in science and statistics. Getting it right helps us find the truth!

---

### **1. What is Bias?**

**Definition 1:** Bias is a **systematic error** that makes our results different from the true, real-world value. Think of it like a **leaning scale**. If a scale always shows 1kg extra, it's not random; it's *systematically* wrong. That's bias.

__Definition 2:__ Any factor that causes a systemetic distortion in data collection.

**Simple Explanation:** It's a consistent push in one direction that makes our findings unfair or inaccurate. It doesn't mean you're being bad on purpose; it often happens by accident because of how we set up our study or ask our questions.

**Edge Case:** Sometimes, what looks like bias might just be a small, random mistake. True bias is a *pattern* of error, not a one-time oopsie.

**Example 1: The Leaning Scale**
Imagine you have a kitchen scale to weigh flour for baking. If something is stuck under it, making every measurement 10 grams heavier, that's bias. Your results are consistently wrong.

**Example 2: Heart Attack Medicine**
A company tests a new tablet to prevent heart attacks. They only test it on very young, healthy athletes. The tablet seems to work great! But this is biased. The results are wrong because they didn't test it on older people who actually get heart attacks. The "scale is leaning" in favor of the medicine.

__Example 3: Black Glasses__
if you see something in black glasses you will see everything is black but actually it is `bias` because the thing is not black in real but base on the glasses it seen like it is black.

> __Bias affects: the Accuracy of data and Reliability of result which mislead to draw conclusion.__

**Uses:** Understanding bias is used to **design better experiments, create fairer surveys, and build more accurate machine learning models.** It's our main tool for finding the truth.

**Benefits of Understanding Bias:**
*   Helps us get closer to the true answer.
*   Makes science more trustworthy.
*   Helps create fair and equitable systems (like fair lending or hiring).

**Limitation:** It can be very tricky to spot and remove *all* bias from a study. Sometimes we only discover it after the work is done.

### **2. Types of Bias in Data**

Bias can sneak into our work in many different ways. Here are the most common types.

#### **1. Sampling Bias**

**Definition:** This happens when the **group you are studying (your sample)** does not properly represent the **larger group you want to understand (the population)**.

__Definition 2:__ Sampling Bias occurs when a sample is not representitive of a population

**Simple Explanation:** It's like tasting a spoonful of soup to see if it's good. If you only taste from the top, you might miss the vegetables at the bottom! Your "sample" (the top) isn't representative of the whole pot.

**Example 1:**
If you want to know the favorite music of everyone in your school, but you only ask people in the band club, your results will be biased. Your sample (band club) doesn't represent the population (the whole school).

#### **2. Selection Bias**


**Definition 1:** This is a type of sampling bias where the way you **select** your participants causes the bias.

__Definition 2:__ This happens when the process of selecting participants or data points affect __the study Outcome__.

**Simple Explanation:** You are picking people for your study in a way that leaves out an important group.

**Example 1:**
Using the **heart attack tablet** example again: If researchers only select young, healthy people for their drug trial (because they are easier to track for 10 years), they have *selected* a biased group. They won't know if the drug works on the sick, older people who need it.

#### **3. Confirmation Bias**

**Definition:** This is the tendency to **favor, search for, or remember information** that confirms what you already believe.

__Definition 2:__ 
- Gut feeling
- Pre-Sumptions
- Pre-Existing Belief

**Simple Explanation:** You see what you want to see. You only pay attention to things that agree with you and ignore things that don't.

**Example 1:**
A scientist believes a certain chemical makes plants grow faster. She might unconsciously give more water or sunlight to the plants with the chemical and ignore the ones without it. She's making the results confirm her belief.

#### **4. Survival Bias**

**Definition 1:** This happens when you only focus on the things or people that "survived" a process and **ignore those that did not**.

__Definition 2:__ 
- You only select or discuss survivors/Sucess stories
- Igoring The failed one

**Simple Explanation:** You only look at the winners and forget about the losers, which gives you a super optimistic view.

**Famous Example 1: World War II Planes**
During the war, engineers looked at bullet holes in planes that returned from battle. They wanted to add armor to the most-hit areas (wings, body). A smart statistician said **NO!** This is survival bias. We are only seeing the planes that *survived* being hit in those areas. The planes that got hit in the engine didn't come back! So, we should armor the engine, the place with *no bullet holes* in the data.

__Example 2: Titanic Ship__

#### **5. Reporting Bias (or Information Bias)**

**Definition:** This occurs when the **way data is collected or reported** is inaccurate. People might not tell the truth, or the method of measuring is faulty.

**Simple Explanation:** The data itself is wrong because people lied, forgot, or the measuring tool was broken.

**Example:**
If you ask students, "How many hours did you *really* study for the test?" many might lie and say a higher number to look good. The *reported* data is biased and doesn't show the true study time.

#### **6. Recall Bias (A type of Information Bias)**

**Definition:** This is when people don't remember past events accurately.

**Simple Explanation:** Your memory is fuzzy, so your answers are biased.

**Example:**
After a school gets sick from bad food, researchers ask students, "What did you eat for lunch yesterday?" Students who got sick are more likely to *remember and report* eating the suspect food, even if they didn't. Healthy students might forget. The memory of the event is biased by the outcome (getting sick).

### __1. Inferential Statistics__
Inferencial Statistics is the branch of statstics that involves using data from a sample to make generalization, predictions or inference about a population.

It relies on __Probability theory__ to estimate population parameters, test hypothses, and draw conclusions beyond the data directly observed.
It's actual the prediction phase.

### __Explanation:__
Inferential statistics goes beyond simply describing data(as in descriptive statistics). Instead, it answers questions like:
- "What does this sample tell us about the whole population?"
- "Is this result likely due to chance?"
- "Can we predict future outcomes based on this data?"

It Uses tools like:
- __Confidence intervals__ -- to estimate population values
- __Hypothesis testing  -- to assess theories
- __Regression analysis(ML) -- to model relationships and make predictions.

__Example__
A researcher surveys 1000 people about their mobile phone usage and finds that about 70% use the smartphones.Using inferential statistics, they predict that about 70% of the entire population uses smartphones -- even though they didn't survey everyone.

__Key Points:__
- Based on samples, not fully populations.
- Involves uncertainty abd probability
- Common in Science, market research, politics and machine learning
