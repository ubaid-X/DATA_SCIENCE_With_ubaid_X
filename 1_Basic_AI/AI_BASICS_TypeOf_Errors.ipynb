# DS vs AI
# 1. Pehle Pehle... Kya Hai Data Science?
Data Science ek buzurg ka ilm hai, jahan hum bohot saari data ko analyze karte hain, use samajhte hain, aur isse valuable
insights nikalte hain. Iska use hota hai business decisions, predictions, aur recommendations mein. Think about it jaise ke ek
biryani ki recipe. Har ingredient ko sahi tarah mix karne par hi aapko behtareen biryani milegi.

 # 2. Ab AI Ki Bari... Kya Cheez Hai Yeh?
AI, yaani Artificial Intelligence. Yeh hai woh masala jo machines ko human jaisa bana deta hai. Matlab, machine ko samajh aata
hai kuch tasks jo pehle sirf insaan kar paate the. AI mein learning models, neural networks jaise advanced tools ka use hota hai.

 # 3. Toh Kahan Aata Hai Connection In Dono Mein?
Socho, Data Science se aapko pata chala ke aapki biryani ki sale har Friday ko zyada hoti hai. Ab, AI use kar ke, aap yeh predict
kar sakte hain ke agle Friday ko kitni biryani ki zaroorat hogi. Samajh rahe ho? Ek insights deta hai, aur doosra uspe action leta
hai.

# 4. Data: AI Ka Khana-Peena
AI algorithms ko chalane ke liye data ki zarurat hoti hai. Aur yeh data kahan se aata hai? Ji haan, Data Science se! Data Scientists
data ko clean, prepare, aur transform karte hain takay AI models isse "learn" kar sakein.

# 5. Machine Learning: The Bridge
Machine Learning (ML) ek subset hai AI ka. Yeh woh jagah hai jahan Data Science aur AI dono milte hain. ML models ko train kiya
jata hai using data, and yeh models predict ya classify karte hain based on this training.

# 6. Real-Life Example Time!
Socho, Lahore mein ek khoobsurat sa shopping mall khula hai. Ab woh chahte hain ke woh samjhein ke log kis waqt sabse zyada
aate hain shop karne. Yeh jaan'ne ke liye, woh data collect karte hain (Data Science). Phir, unhone ek AI system develop kiya jo
predict karta hai ke agle hafte kis din aur kis waqt log zyada aayenge (AI). Aha! Dekha? Dono hand-in-hand chal rahe hain.

# 7. Ek Aur Masaledar Example!
Maan lo aapke paas ek restaurant hai jahan har roz nayi-nayi dishes try ki jaati hain. Aap Data Science use kar ke pata laga lete
hain ke log gosht ki konsi dish zyada pasand karte hain. Phir, AI ka magic use hota hai predicting ke liye ke agle mahine Eid par
konsi dish zyada bikne wali hai.
Data science bootcamps Data science courses

# 8. Future: AI aur Data Science ka Milan
Future mein, humein expect karna chahiye ke AI aur Data Science ka milan aur bhi zyada hoga. Personalized experiences, smart
cities, aur health predictions, yeh sab kuch in dono ke combination se hi possible hoga.

# 9. Rishte Mein Toh Hum...
Toh dosto, humara conclusion kya hai? Simple: Data Science aur AI judwa bhaai nahi, par woh hain best friends jo ek dusre ke
bina adhoore hain. Data Science woh foundation hai jis par AI ka mahal khada hota hai.

# 10. Aakhri Alfaaz

# types of error in python

# 1.SyntaxError
Cause: Invalid Python syntax.
Example:


if True print("Hello")
Fix: Missing : ‚Üí if True: print("Hello")


# 2.NameError
Cause: Using a variable that hasn't been defined.
Example:


print(x)
Fix: Define x first ‚Üí x = 5


# 3.TypeError
Cause: Incompatible data types.
Example:


print("Age: " + 20)
Fix: Convert integer ‚Üí print("Age: " + str(20))


# 4.ValueError
Cause: Invalid value for an operation.
Example:

int("abc")
Fix: Ensure convertible string ‚Üí int("123")


# 5.IndexError
Cause: Accessing out-of-range list index.
Example:

lst = [1, 2] print(lst[5])
Fix: Use valid index ‚Üí print(lst)

# 6.KeyError
Cause: Accessing missing key in dictionary.
Example:

d = {"a": 1} print(d["b"])
Fix: Use .get() ‚Üí d.get("b")


# 7.AttributeError
Cause: Calling an attribute/method that doesn‚Äôt exist.
Example:

x = 5 x.append(3)
Fix: Use method on right object type.


# 8.ZeroDivisionError
Cause: Dividing by zero.
Example:

print(10 / 0)


# 9.FileNotFoundError
Cause: File path doesn‚Äôt exist.

# 10.ImportError
