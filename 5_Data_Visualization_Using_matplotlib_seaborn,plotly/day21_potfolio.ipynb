> # Steps to create a portfolio

1. Provide a Catchy Introduction
2. Add a More Thorough "About" Section
3. Include an Easy Way to Contact you(not whatsapp ,not improfesional)
4. Add your most promising Project(Less Is more):
   1. Highlight work done for Prominent Companies in your Niche
   2. Focus on Common Projects (solve Common Problem)
   3. Consider Adding a page or Section for Each Prject
   4. Include Project Detail, Learning and Takeaways
5. Make it easy to navigate(sample and Concise)
6. How to Stand Out:
   1. Use a Self-Hosted Solution
   2. Invest in an Apealing Design and layout
   3. Talk About the types of Projects that interest you
   4. Illustrate work trait that will help Potential Client Get to know you
   5. Document the tools you're Familiar with and how you've use them in the past
7. What to Avoid
   1. Never Copy Another Portfolio
   2. Never use Formulatic Website Templates
   3. Don't Include Projects you aren't Proud of
