> # Plotting using seaborn

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Important things for plotting according to grammer of grafics

1. Data (df)
2. Types of plot?
3. y-Axis
4. X-axis
5. for Grouping (hue / color) based on categorical variable
6. Size of the plot
7. color palletes of the plot
8. style of the plot
9. How do you interpret

```python
# loading data
df = pd.read_csv('../5_Datasets/titanic.csv')
```

```python
df.info()
```

### Step-3,4 => Size and color palette of plot

```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(8, 4))

sns.scatterplot(data = df, x='age', y='fare', hue = 'sex')
```

### change the scatter point colors manually

```python

# scatter plot size of figure size
plt.figure(figsize=(8, 4))

# change the scatter point colors
palette = {'male': 'blue', 'female': 'red'}

sns.scatterplot(data = df, x='age', y='fare', hue = 'sex', palette=palette)

# labels + title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')

# saving a plot 300dpi normal, 600dpi for huge screen, png
plt.savefig('./viz outputs/plots/scatter_Agefare.png', dpi = 300,  bbox_inches = 'tight')

# saving a plot in vector file such as svg
plt.savefig('./viz outputs/plots/scatter_plot.svg')
```

### style of plot

```python
# Style of a plot
sns.set_theme(style='darkgrid', font_scale=1.5, font='calibri')
# scatter plot size of figure size
plt.figure(figsize=(8, 4))

# change the scatter point colors
palette = {'male': 'blue', 'female': 'red'}

sns.scatterplot(data = df, x='age', y='fare', hue = 'sex', palette=palette)

# labels + title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
```

 __Interpretation:__

 1. We have Age of passengers at x-Axis and fare at y-Axis.
 2. The scatter plot shows the relationship between age and fare of titanic passengers.
 3. We further grouped the passenger by their genders such as male (Blue points) and female (Red Points).
 4. Two male passengers gave the highest fare in titanic, and they might be traveling in first class.
 5. The oldest person at titanic is also a male having 80 year of age as shown in the scatter plot. (As shown in Figure. 1)

> # Box Plot

```python
# Style of a plot
sns.set_theme(style='darkgrid', font_scale=1.3, font='calibri')
# scatter plot size of figure size
plt.figure(figsize=(12, 6))

# change the scatter point colors
palette = {'First': 'blue', 'Second': 'pink', 'Third': 'green'}

sns.boxplot(data = df, x='sex', y='age', hue = 'class', palette=palette)

# labels + title
plt.xlabel('sex')
plt.ylabel('Age')
plt.title('Boxen plot of age, based on sex and class name of passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')
```

> # Boxen Plot

```python
# Style of a plot
sns.set_theme(style='darkgrid', font_scale=1.3, font='calibri')
# scatter plot size of figure size
plt.figure(figsize=(12, 6))

# change the scatter point colors
palette = {'First': 'blue', 'Second': 'pink', 'Third': 'green'}

sns.boxenplot(data = df, x='sex', y='age', hue = 'class', palette=palette)

# labels + title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of age, based on gender and class name of passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x ticks labels
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
```

> # bar plot using seaborn 

```python
# Style of a plot
sns.set_theme(style='darkgrid', font_scale=1.3, font='calibri')
# scatter plot size of figure size
plt.figure(figsize=(12, 6))

# change the scatter point colors
palette = {'First': 'green', 'Second': 'blue', 'Third': 'pink'}

sns.barplot(data = df, x='sex', y='age', hue = 'class', palette=palette, capsize=0.1) # set cap size

# labels + title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of age, based on gender and class name of passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x ticks labels
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])


```

> # violin plot

```python
# Style of a plot
sns.set_theme(style='darkgrid', font_scale=1.3, font='calibri')
# scatter plot size of figure size
plt.figure(figsize=(12, 6))

# change the scatter point colors
palette = {'First': 'green', 'Second': 'blue', 'Third': 'pink'}

sns.violinplot(data = df, x='sex', y='age', hue = 'class', palette=palette) # set cap size

# labels + title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title(' violin plot of age, based on gender and class name of passenger')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x ticks labels
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])


```
