# 1. Sunburst Plot to show heirarchy

```python
import plotly.express as px
import seaborn as sns
```

```python
# Load titanic dataset
df = sns.load_dataset('titanic')
```

```python
# creating sunburst plot on titanic dataset
fig = px.sunburst(df, path=['sex', 'class', 'who', 'alive', 'alone'],  color='sex')
fig.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

> Adding more values when we hover the pointer on the plot in sunburst plot 

```python
# Convert categorical columns to string to avoid TypeError
tips_sunburst = tips.copy()
for col in ['sex', 'smoker', 'day', 'time']:
    tips_sunburst[col] = tips_sunburst[col].astype(str)

grouped = tips_sunburst.groupby(['sex', 'smoker', 'day', 'time']).agg({'size': 'sum',
                                                                    'tip': 'sum',
                                                                    'total_bill': 'sum'}).reset_index()

fig1 = px.sunburst(
    grouped,
    path=['sex', 'smoker', 'day', 'time'],
    values='size', hover_data=['tip', 'total_bill'],
    title='Sunburst Plot of Tips Dataset',
    color='smoker',  # Color by smoker status
)
fig1.show()
```

> # saving plot as .gif animation

```python
import plotly.express as px 
import pandas as pd 
import numpy as np 
import io
from PIL import Image
```

```python
# Animated scatter bubble chart
fig = px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent',
                 animation_frame='year', animation_group='country',
                 log_x=True, size_max=55, range_x=[100, 100000], range_y=[25, 90])

# generate image for each step in animation
frames = []
for s, fr in enumerate(fig.frames):
    # set main traces to appropriate traces within plotly frame
    fig.update(data=fr.data)
    # move slider to correct place
    fig.layout.sliders[0].update(active=s)
    # generate image of current state
    frames.append(Image.open(io.BytesIO(fig.to_image(format='png', scale=3))))

# create animated GIF
frames[0].save(
    './viz outputs/plots/gapminder.gif',
    save_all=True,
    append_images=frames[1:],
    optimize=True, 
    duration=500,
    loop=0,
    dither=None # turn off dithering
)
```

> # Geo Maps    

```python
import plotly.express as px
df = px.data.election()

geojson = px.data.election_geojson()   # base map
print(df.head())
print(df['district'][2])
print(geojson['features'][0]['properties'])
```

```python
fig = px.choropleth_map(df, geojson=geojson, color='Joly',
                           locations='district', featureidkey='properties.district',
                           center={'lat': 45.5517, 'lon': -73.7073},
                           map_style='carto-positron', zoom=9)
fig.update_layout(margin={'r':0, 't':0, 'l':0, 'b':0})
fig.show()
```

```python
import pandas as pd
# sample population data country wise
data = {
    'country': ['china', 'india','USA', 'pakistan', 'Canada', 'Mexico', 'Germany', 'France',],
    'population': [1400000000, 1380004385, 331002651, 225199937, 37742154, 128932753, 83783942, 65273511],
} 

# create a pandas dataframe
df = pd.DataFrame(data)

# Display the Dataframe
print(df)
```

```python
import plotly.express as px
# create a choropleth map using plotly express 
fig = px.choropleth(df, locations='country', locationmode='country names',
                           color='population',
                           color_continuous_scale=px.colors.sequential.Plasma,
                           title='Population by Country')
fig.show()
```

```python
# create a choropleth map with animation using plotly express 
fig = px.choropleth(df, locations='country', locationmode='country names',
                           color='country', animation_frame='population',
                           color_continuous_scale=['red', 'orange', 'yellow', 'green', 'blue'],
                           title='Population by Country', )
fig.show()
```
