✅ 4. *Correlation Matrix*

---

🔹 What is Correlation?

*Correlation* tells *how strongly two numeric columns are related*.

- Value ranges from *-1 to +1*
  - *+1* → Strong positive relation (both increase together)
  - *0* → No relation
  - *-1* → Strong negative relation (one increases, other decreases)

---

🔹 Real Example

Imagine a dataset of students:

```python
import pandas as pd

data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Marks':         [40, 50, 60, 70, 80],
    'Sleep_Hours':   [8, 7, 6, 5, 4]
}

df = pd.DataFrame(data)
print(df)
```

*DataFrame:*

| Hours_Studied | Marks | Sleep_Hours |
|---------------|-------|-------------|
| 1             | 40    | 8           |
| 2             | 50    | 7           |
| 3             | 60    | 6           |
| 4             | 70    | 5           |
| 5             | 80    | 4           |

---

🔹 Apply Correlation

```python
correlation = df.corr()
print(correlation)
```

*Output (Correlation Matrix):*

|               | Hours_Studied | Marks | Sleep_Hours |
|---------------|---------------|-------|--------------|
| Hours_Studied | 1.00          | 1.00  | -1.00        |
| Marks         | 1.00          | 1.00  | -1.00        |
| Sleep_Hours   | -1.00         | -1.00 | 1.00         |

---

🔹 Easy Interpretation

- `Hours_Studied` & `Marks` → *+1.00* = perfect positive relation
- `Sleep_Hours` & `Hours_Studied` → *-1.00* = one goes up, other down
- Diagonal is always `1` (correlation with itself)
